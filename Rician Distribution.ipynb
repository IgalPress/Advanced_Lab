```python
import numpy as np
import matplotlib.pyplot as plt

#generate and plot normally distributed randoms numbers about zero






npts=100000

mean=np.sqrt(0.0)
sigma=1

rnum1=sigma*np.random.randn(npts) + mean
rnum2=sigma*np.random.randn(npts) + mean

mean1=np.round(np.mean(rnum1),3)
stddev1=np.round(np.std(rnum1),3)
mean2=np.round(np.mean(rnum2),3)
stddev2=np.round(np.std(rnum2),3)


print('The mean is %s and the standard deviation is %s.'  % (mean1, stddev1))
print('The mean is %s and the standard deviation is %s.'  % (mean2, stddev2))
print()


mag=[]
for i in range(0,npts):
    mag.append(np.sqrt(rnum1[i]**2+rnum2[i]**2))



plt.hist(rnum1,51)
plt.grid(True)
plt.xlabel('Random number')
plt.ylabel('Counts')
plt.title('The normal distribution #1')

plt.figure()

plt.hist(rnum2,51)
plt.grid(True)
plt.xlabel('Random number')
plt.ylabel('Counts')
plt.title('The normal distribution #2')

plt.figure()

plt.hist(mag,50)
plt.grid(True)
plt.xlabel('Random number')
plt.ylabel('Counts')
plt.title('The Rician distribution')


plt.figure()

y,binEdges=np.histogram(mag,50)
bincenters = 0.5*(binEdges[1:]+binEdges[:-1])
plt.plot(bincenters,y,'-')

plt.show()
```
