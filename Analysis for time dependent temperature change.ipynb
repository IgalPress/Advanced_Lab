```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df2 = pd.read_csv('Min_Max.csv', header = None)
#plt.figure(figsize=(9,6),dpi=200)
plt.grid()
plt.xlabel('Time (Mins)')
plt.ylabel('Temperature (C)')
plt.plot(df2[0], df2[1], label = 'Temperature Increase')
plt.legend()
p = np.polyfit(df2[0], df2[1], 1)
print(p[1])
print(np.exp(p[1]))
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np

# Sample data
x = np.array(df2[0]).reshape(-1, 1)
y = np.array(df2[1])

# Create linear regression model
model = LinearRegression()

# Fit model to data
model.fit(x, y)

# Print the coefficients
print('Intercept:', model.intercept_)
print('Slope:', model.coef_)
print(model.predict(x))

x_test = np.array([5]).reshape(-1, 1)
y_pred = model.predict(x_test)
print("Predicted y value at x = 10:", y_pred[0])
```

```python
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt

mins = df2[0]

def capacitor_charge(t, a, tau, c):
    return a * (1 - np.exp(-t/tau)) + c



popt, pcov = curve_fit(capacitor_charge, df2[0], df2[1])

minfit = np.linspace(0, 61, 1)

print(popt)

minfit = np.linspace(0, 63, 64)

plt.plot(minfit, capacitor_charge(minfit, *popt), 'r-', label='fit')

plt.scatter(mins, df2[1])
plt.title('0 - max')
plt.plot(mins - 1, capacitor_charge(mins - 1, *popt), 'r-')
plt.legend()
plt.xlabel("mins")
plt.ylabel('temp')
plt.grid()
plt.show()
```
