```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression

df = pd.read_csv('dataleslie.csv')
df
dull = df['Dull']
black = df['Black']
shiny = df['Shine']
white = df['White']
T4 = df['T^4']
plt.scatter(T4, black, label = 'Black')
plt.scatter(T4, white, label = 'White')
plt.scatter(T4, dull, label = 'Dull')
plt.scatter(T4, shiny, label = 'Shiny')
plt.grid()
plt.legend()
plt.xlabel('$T^4$')
plt.ylabel('Radiation (mV)')
from scipy import stats
slope, intercept, r, p, std_err = stats.linregress(T4, dull)
slope2, intercept2, r2, p2, std_err2 = stats.linregress(T4, black)
slope3, intercept, r, p, std_err = stats.linregress(T4, white)
slope4, intercept2, r2, p2, std_err2 = stats.linregress(T4, shiny)
def predict85(x):
    return slope*x+intercept
def predict87(x):
    return slope2*x+intercept2
def predict88(x):
    return slope3*x+intercept2
def predict89(x):
    return slope4*x+intercept2
mymodel = list(map(predict85, T4))
mymodel2 = list(map(predict87, T4))
mymodel3 = list(map(predict88, T4))
mymodel4 = list(map(predict89, T4))
plt.plot(T4, mymodel, color = 'green')
plt.plot(T4, mymodel2, color = 'blue')
plt.plot(T4, mymodel3, color = 'orange')
plt.plot(T4, mymodel4, color = 'red')
plt.plot()
plt.savefig('radvT.png', bbox_inches='tight')
```
