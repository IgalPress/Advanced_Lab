```python
from scipy import stats
slope, intercept, r, p, std_err = stats.linregress(df3[0], df3[1])
slope2, intercept2, r2, p2, std_err2 = stats.linregress(df3[0], df3[2])
def predict85(x):
    return slope*x+intercept
def predict87(x):
    return slope2*x+intercept2
mymodel = list(map(predict85, df3[0]))
mymodel2 = list(map(predict87, df3[0]))
plt.plot(df3[0], mymodel, label = '$^{85}$Rb line of best fit, slope =  2.309(1)e-03 A/Hz')
plt.plot(df3[0], mymodel2, label = '$^{87}$Rb line of best fit, slope = 3.445(1)e-03 A/Hz')
plt.errorbar( df3[0], df3[1] , yerr=std_err, fmt='.', label = 'Data for $^{85}$Rb' )
plt.errorbar( df3[0], df3[2] , yerr=std_err2, fmt='.', label = 'Data for $^{87}$Rb' )
print('The values for 85Rb are: slope =', slope,4, 'intercept =', np.round(intercept,4), 'r^2 =', r**2, 'std_err =', np.round(std_err,10))
print('The values for 87Rb are: slope =', slope2, 'intercept =', np.round(intercept2,4), 'r^2 =', r2**2, 'std_err =', np.round(std_err2,10))
prediction85 = predict85(5000)
prediction87 = predict87(5000)
print('The value of the current running through the horizontal helmholtz coil at frequency xHz for 85Rb is', prediction85)
print('The value of the current running through the horizontal helmholtz coil at frequency xHz for 87Rb is', np.round(prediction87,4))
plt.ylabel('Current in Helmholtz coil (A)')
plt.xlabel('Frequency in function generator (Hz)')
print('error between the two slopes compared to 2/3 is: ', ((slope/slope2)*100, '%'))
plt.legend(fontsize = 'x-small')
plt.grid(linestyle='--', linewidth = 0.7)
plt.savefig('LFRefficiency.png')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np

# Sample data
x = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
y = np.array([5, 7, 9, 11, 13])

# Create linear regression model
model = LinearRegression()

# Fit model to data
model.fit(x, y)

# Print the coefficients
print('Intercept:', model.intercept_)
print('Slope:', model.coef_)
print(model.predict(x))

x_test = np.array([5]).reshape(-1, 1)
y_pred = model.predict(x_test)
print("Predicted y value at x = 10:", y_pred[0])
```
