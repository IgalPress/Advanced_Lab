```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# Bloch equation with no relaxation
# Function returns the three components of dM/dt
def Bloch(M, t, f):
    x = M[0]
    y = M[1]
    z = M[2]
    dxdt = 2 * np.pi * f * y
    dydt = - 2 * np.pi * f * x
    dzdt = 0
    dMdt = [dxdt, dydt, dzdt]
    return dMdt


# initial conditions
M0 = [1, 0, 0]  # Mx =1, My =0, Mz =0

# number of time points
n = 1001

# time points
t = np.linspace(0, 4, n)


# solve ODE
freq = 4
Mt = odeint(Bloch, M0, t, args=(freq,))


# plot results
plt.plot(t, Mt[:, 0], 'b-', label='Mx')
plt.plot(t, Mt[:, 1], 'r-', label='My')
plt.plot(t, Mt[:, 2], 'g-', label='Mz')
plt.axis([-0.2, 4.2, -1.2, 1.2])
plt.ylabel('Magnetization')
plt.xlabel('Time (s)')
plt.legend(loc='best')
#plt.savefig("q3a.png")
plt.show()
```
