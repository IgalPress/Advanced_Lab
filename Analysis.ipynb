```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

dfsamp96 = pd.read_csv('Penny96.csv', sep = ',') #Penny96
dfsamp99 = pd.read_csv('Penny99.csv', sep = ',') #Penny99
dfsamp031 = pd.read_csv('Penny2003-1.csv', sep = ',') #Penny2003-1
dfsamp032 = pd.read_csv('Penny2003-2.csv', sep = ',') #Penny2003-2

dfsamp64 = pd.read_csv('Silver64.csv', sep = ',') #Silver64
dfsamp67 = pd.read_csv('Silver67.csv', sep = ',') #Silver67
dfsamp68 = pd.read_csv('Silver67.csv', sep = ',') #Silver68
dfsamp69 = pd.read_csv('Silver69.csv', sep = ',') #Silver69

dfnorm = pd.read_csv('NormIn.csv', sep = ',' , skiprows = 3 )

phase_norm = dfnorm['PHASE M->4']
magnitude_norm = dfnorm['VAmplitude']

#Penny96
phase_sample96 = dfsamp96['PHASE M->4']
magnitude_sample96 = dfsamp96['VAmplitude']
phase96_corr = phase_sample96 - phase_norm
mag96_corr = magnitude_sample96 / magnitude_norm

#Penny99

phase_sample99 = dfsamp99['PHASE M->4']
magnitude_sample99 = dfsamp99['VAmplitude']
phase99_corr = phase_sample99 - phase_norm
mag99_corr = magnitude_sample99 / magnitude_norm

#Penny2003-1

phase_sample031 = dfsamp031['PHASE M->4']
magnitude_sample031 = dfsamp031['VAmplitude']
phase031_corr = phase_sample031 - phase_norm
mag031_corr = magnitude_sample031 / magnitude_norm

#Penny2003-2

phase_sample032 = dfsamp032['PHASE M->4']
magnitude_sample032 = dfsamp032['VAmplitude']
phase032_corr = phase_sample032 - phase_norm
mag032_corr = magnitude_sample032 / magnitude_norm

#Silver64

phase_sample64 = dfsamp64['PHASE M->4']
magnitude_sample64 = dfsamp64['VAmplitude']
phase64_corr = phase_sample64 - phase_norm
mag64_corr = magnitude_sample64 / magnitude_norm

#Silver67

phase_sample67 = dfsamp67['PHASE M->4']
magnitude_sample67 = dfsamp67['VAmplitude']
phase67_corr = phase_sample67 - phase_norm
mag67_corr = magnitude_sample67 / magnitude_norm

#Silver68

phase_sample68 = dfsamp68['PHASE M->4']
magnitude_sample68 = dfsamp68['VAmplitude']
phase68_corr = phase_sample68 - phase_norm
mag68_corr = magnitude_sample68 / magnitude_norm

#Silver67

phase_sample69 = dfsamp69['PHASE M->4']
magnitude_sample69 = dfsamp69['VAmplitude']
phase69_corr = phase_sample69 - phase_norm
mag69_corr = magnitude_sample69 / magnitude_norm
```

```python
plt.title('Penny Mag')
plt.grid()
plt.ylabel('Magnitude')
plt.xlabel('Frequency (Hz)')
plt.xscale('log')
plt.plot(dfsamp96['%Frequency'], mag96_corr, label = 'penny 1996')
plt.plot(dfsamp96['%Frequency'],mag99_corr, label = 'penny 1999')
plt.plot(dfsamp96['%Frequency'],mag031_corr, label = 'penny 2003-1')
plt.plot(dfsamp96['%Frequency'],mag032_corr, label = 'penny 2003-2')
plt.legend()
```

```python
plt.title('Penny Phase')
plt.grid()
plt.ylabel('Phase difference')
plt.xlabel('Frequency (Hz)')
plt.xscale('log')
plt.plot(dfsamp96['%Frequency'], phase96_corr, label = 'penny 1996')
plt.plot(dfsamp96['%Frequency'],phase99_corr, label = 'penny 1999')
plt.plot(dfsamp96['%Frequency'], phase031_corr, label = 'penny 2003-1')
plt.plot(dfsamp96['%Frequency'],phase032_corr, label = 'penny 2003-2')
plt.legend()
```

```python
plt.title('Silver Phase')
plt.grid()
plt.xlabel('Phase difference')
plt.ylabel('Frequency (Hz)')
plt.plot(dfsamp96['%Frequency'], phase64_corr, label = 'silver 66')
plt.plot(dfsamp96['%Frequency'], phase67_corr, label = 'silver 67')
plt.plot(dfsamp96['%Frequency'], phase68_corr, label = 'silver 68')
plt.plot(dfsamp96['%Frequency'], phase69_corr, label = 'silver 69')
```

```python
plt.title('Silver mag')
plt.grid()
plt.xlabel('Phase difference')
plt.ylabel('Magnitude (V)')
plt.plot(mag64_corr, dfsamp96['%Frequency'])
plt.plot(mag67_corr, dfsamp96['%Frequency'])
plt.plot(mag68_corr, dfsamp96['%Frequency'])
plt.plot(mag69_corr, dfsamp96['%Frequency'])
```
